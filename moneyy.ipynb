```python
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
```

```python
opa = yf.Ticker('PETR4.SA')
data_hoje = dt.date.today()
base = opa.history(start='2023-01-01',end=data_hoje)
base2 = opa.history(period='3mo')
base3 = opa.history(period='1mo')
```

```python
figura = plt.figure(figsize=(40,6))
figura.suptitle('PETR4')
figura.add_subplot(131)

plt.plot(base.index,base.Close, label= opa.ticker,lw="3")
plt.xlabel('Periodo')
plt.ylabel('Preço')
plt.title('Começo do ano')
plt.legend()

figura.add_subplot(132)
plt.plot(base2.index,base2.Close, lw="3")
plt.title('6 meses')

figura.add_subplot(133)
plt.plot(base3.index,base3.Close, lw="3")
plt.title('Último mês')

plt.savefig('Gráficos.png')
```

```python
import plotly.express as px
fig = px.line(base,x=base.index,y=base.Close)
fig.show()
```
